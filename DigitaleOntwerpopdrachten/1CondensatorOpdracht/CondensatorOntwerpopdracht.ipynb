Zie probleem 2.6 in het boek van Griffith.

##### Antwoord 2h

$$
E_{\text{schijf}}=0.1\,E_{\infty}
$$
invullen geeft:
$$
\frac{\sigma}{2\epsilon_0}
\left(1-\frac{z}{\sqrt{R^2+z^2}}\right)
=
0.10  \frac{\sigma}{2\epsilon_0}
$$
de factor $\frac{\sigma}{2\epsilon_0}$ kan worden weggestreept:
$$
\left(1-\frac{z}{\sqrt{R^2+z^2}}\right)
=
0.10
$$

$$
\frac{z}{\sqrt{R^2+z^2}} = 0.90
$$
kwadrateren geeft het volgende:
$$
\frac{z^2}{R^2+z^2} = 0.81
$$
verder oplossen geeft:
$$
z^2 = 0.81(R^2+z^2)
$$

$$
0.19 z^2 = 0.81 R^2
$$

$$
z = R \sqrt{\frac{0.81}{0.19}}
$$
